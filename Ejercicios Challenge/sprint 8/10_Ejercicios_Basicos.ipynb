### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(2, 2, figsize=(8, 6))  # Tamaño de la figura ajustado

# Graficar en cada subplot
ax[0, 0].plot(x, np.sin(x))
ax[0, 0].set_title('Seno')

ax[0, 1].plot(x, np.cos(x))
ax[0, 1].set_title('Coseno')

ax[1, 0].plot(x, np.tan(x))
ax[1, 0].set_title('Tangente')

ax[1, 1].plot(x, -np.sinh(x))
ax[1, 1].set_title('Seno hiperbólico negativo')

# Ajustar el espaciado entre subplots
fig.subplots_adjust(hspace=0.5, wspace=0.4)

plt.show()
fig
```
